```python
import numpy as np
import pandas as pd
import requests
import json
 
#请求地址
url = " https://api.spacexdata.com/v4/launches/past"
response = requests.get(url)

response=response.json()
```

```python
import csv
import json
import sys
import codecs

jsonData = response
csvfile = open('C:/Users/ROG/Desktop/DATA7901/luanch.csv', 'w', newline='')  # python3下
writer = csv.writer(csvfile, delimiter=',')
keys = ['FlightNumber','rocketid','Date','Flights', 'GridFins', 'Reused', 'Legs', 
        'LandingPad', 'Outcome', 'LandingType','Coresid','payloadsid']
writer.writerow(keys)

for line in jsonData:
    cores=line['cores']
    Flights=cores[0]['flight']
    GridFins=cores[0]['gridfins']
    Reused=cores[0]['reused']
    Legs=cores[0]['legs']
    Outcome=cores[0]['landing_success']
    LandingType=cores[0]['landing_type']
    Coresid=cores[0]['core']
    writer.writerow([line['flight_number'],line['rocket'],line['date_utc'],Flights,GridFins,Reused,Legs,
                     line['launchpad'],Outcome,LandingType,Coresid,line['id']]
                     )
csvfile.close()
    
```

```python
luanch=pd.read_csv('C:/Users/ROG/Desktop/DATA7901/luanch.csv')
cores=pd.read_csv('C:/Users/ROG/Desktop/DATA7901/cores.csv')
launchpads=pd.read_csv('C:/Users/ROG/Desktop/DATA7901/luanchpad.csv')
payloads=pd.read_csv('C:/Users/ROG/Desktop/DATA7901/Payloads.csv')
rocket=pd.read_csv('C:/Users/ROG/Desktop/DATA7901/rocket.csv')
```

```python
luanch=luanch.reset_index(drop=True)
cores=cores.reset_index(drop=True)
launchpads=launchpads.reset_index(drop=True)
payloads=payloads.reset_index(drop=True)
```

```python
df_merged=pd.merge(left=luanch,right=cores,on='Coresid')
df_merged=pd.merge(left=df_merged,right=launchpads,on='LandingPad')
df_merged=pd.merge(left=payloads,right=df_merged,on='payloadsid')
df_merged=pd.merge(left=rocket,right=df_merged,on='rocketid')
```

```python
df_merged
```

```python
df_merged1=df_merged.drop(columns='PayloadMass')
df_merged1
```

```python
df_merged1=df_merged1.drop_duplicates('payloadsid',keep='first')
```

```python
df_merged2=df_merged.groupby(['Date'])[['PayloadMass']].sum()
df_merged3=pd.merge(left=df_merged1,right=df_merged2,on='Date')
df_merged3
```

```python
df_merged3=df_merged3.drop(columns=['LandingPad','Coresid','payloadsid'])
```

```python
df_merged3.reset_index(drop=True)
columns = list(df_merged3)
columns.insert(0, columns.pop(columns.index('FlightNumber')))
df_merge4 = df_merged3.loc[:, columns]
df_merge4
```

```python
m=0
for i in df_merge4['Date']:
    a,b=i.split('T')
    print(a)
    df_merge4['Date'][m]=a
    m+=1
```

```python
df_merge4=df_merge4.sort_values(by=['FlightNumber'])
df_merge4['Outcome']=df_merge4['Outcome'].fillna(value='None')
df_merge4['LandingType']=df_merge4['LandingType'].fillna(value='None')
```

```python
df_merge5=df_merge4
df_merge5
```

```python
df_merge4.to_csv('C:/Users/ROG/Desktop/DATA7901/out1.csv')
df_merge5.to_csv('C:/Users/ROG/Desktop/DATA7901/out2.csv')
```
